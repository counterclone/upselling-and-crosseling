```python
import pandas as pd
import math
import random
def calculate_credit_score(income, history_length, credit_balance):
    # Weightage factors (you can adjust these based on real-life scenarios)
    income_weight = 0.4 + random.uniform(-0.05, 0.05)
    history_weight = 0.3 + random.uniform(-0.05, 0.05)
    balance_weight = 0.3 + random.uniform(-0.05, 0.05)

    # Normalize inputs
    normalized_income = (income - 30000) / (5000000 - 30000)  # Annual salary between 30000 and 5000000
    normalized_history = history_length / 50  # History length between 0 and 50 years
    normalized_balance = (credit_balance - 30000) / (5000000 - 30000)  # Credit balance is fixed at 30000

    # Calculate credit score
    credit_score = (
        income_weight * normalized_income +
        history_weight * normalized_history +
        balance_weight * normalized_balance
    )

    # Ensure credit score is within the desired range (200 - 1000)
    #credit_score = max(min(credit_score, 1000), 200)

    return credit_score*1000





```

```python
def calculate_credit_limit(credit_score, annual_salary,min1,max1,min2,max2):
    # Weightage factors (you can adjust these based on real-life scenarios)
    credit_score_weight = 0.6 + random.uniform(-0.05, 0.05)
    salary_weight = 0.4 + random.uniform(-0.05, 0.05)

    # Normalize inputs
    normalized_credit_score = (credit_score - min1) / (max1-min1)  # Credit score between 200 and 900
    normalized_salary = (annual_salary - min2) / (max2 - min2)  # Annual salary between 300000 and 5000000

    # Calculate credit limit
    credit_limit = (
        credit_score_weight * normalized_credit_score +
        salary_weight * normalized_salary
    ) * 10000  # Multiplying by 10,000 for better scale

    return credit_limit*1000
```

```python
def predict_partial_payment(age, annual_salary, credit_score, credit_limitation, credit_card_loan):
    n=0
    t=credit_card_loan
    while(t>1):
      n+=1
      t=math.floor(t/10)
    p=random.randint(0,n-1)
    credit_score-=250
    partial_payment = math.ceil(credit_card_loan/math.pow(10,p))
    if(partial_payment==0):
      return partial_payment
    p=0
    t=partial_payment
    while(t>1):
      p+=1
      t=math.floor(t/10)

    q=0
    t=annual_salary
    while(t>1):
      q+=1
      t=math.floor(t/10)

    w=(age-21)*math.pow(10,p-2)+(annual_salary)*math.pow(10,t-p)
    if(partial_payment>w):
      final= partial_payment-w
    else:
      final= partial_payment

    if(final<100):
      return 0
    else:
      return int(partial_payment)






    # # partial_payment =credit_card_loan/math.pow(10,p)+age*math.pow(10,n-p)-annual_salary*math.pow(10,n-p)*credit_score
    # partial_payment =credit_card_loan/math.pow(10,p)+age*math.pow(10,n-p)

    return partial_payment
```
